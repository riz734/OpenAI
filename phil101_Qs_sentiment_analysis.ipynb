```python
import os
import openai
import pandas as pd
```

```python
def open_ai(prompt):
    response = openai.Completion.create(
      model="text-davinci-003",
      prompt=prompt,
      temperature=0.7,
      max_tokens=64,
      top_p=1.0,
      frequency_penalty=0.0,
      presence_penalty=0.0
    )
    
    return response["choices"][0]["text"].strip()
```

```python
open_ai("Where is U of M located?")
```
